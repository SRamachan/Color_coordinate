This is a simple exercise to specify the color coordinate (specified as random integer between 0-255) for a collection of small dots.

```python
import turtle as t
import numpy as np

# set the 256 bit color scheme
t.colormode(255)
# set-up Tutle screen and background color
screen=t.Screen()
screen.bgcolor("White")
# lift up the pen so there is no trace
t.penup()
# speed of the pen while drawing
t.speed(0)
# shape of the Turtle
t.shape("turtle")
t.goto(-400,400)

# draw five circles
for steps in range(8):
    # select the r,g and b color coordinate at random
    r = np.random.randint(0,255)
    g = np.random.randint(0,255)
    b = np.random.randint(0,255)
    c = [(r,g,b)]
    # print the selected r,g,b color value.
    print("r=", r, ";g=", g, ";b=", b)
    for r, g, b in c:
        # define the color based on selected r,g,b and draw a filled dot of size 
        t.color(r,g,b)
        # draw a dot that is 40 pixels large
        t.dot(40)
        t.write(c)
        # move tutrle forward 150 pixels and 14 degrees below the horizontal axis
        t.forward(150)
        t.right(14)
t.done()
```
