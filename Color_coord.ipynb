This is a simple exercise to specify the color coordinate (specified as random integer between 0-255) for a collection of single dot.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import PatchCollection
from matplotlib.patches import Circle
patches = []
fig, ax = plt.subplots()
R = np.random.randint(0,255)
G = np.random.randint(0,255)
B = np.random.randint(0,255)
r  = R/(R+G+B)
g = G/(R+G+B)
b = B/(R+G+B)
print ("R=", R,";G=", G,";B=", B)
print(f'fractional color, r= {r:.3f};g={g:.3f} ;b={b:.3f}')
cRGB = [R,G,B]
c_RGB_norm = [c/255.0 for c in cRGB]
circ1 = Circle((0.5,0.5),0.25, edgecolor=c_RGB_norm, facecolor=c_RGB_norm)
ax.add_patch(circ1)
ax.set_aspect('equal')
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
plt.show()
```
